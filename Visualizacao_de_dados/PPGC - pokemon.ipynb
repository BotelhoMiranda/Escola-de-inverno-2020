```
df = pd.read_csv('pokemon.csv', index_col=0,encoding='latin-1')
```

```
#Dados de estatísticas de combate para o 151 pokémons originais
df.head(50)
```

```
#Criando um gráfico de dispersão, que é uma maneira padrão de visualizar a relação entre duas variáveis

sns.lmplot(x='Attack', y='Defense', data=df,height=5)
```

Seaborn não tem uma função própria de gráfico de dispersão, é por isso que vemos uma linha diagonal. Na verdade, usamos a função para ajustar e traçar uma linha de regressão. Cada função de plotagem possui várias opções úteis que você pode definir. Veja como podemos ajustar o lmplot ():


```
#it_reg = False para remover a linha de regressão, já que queremos apenas um gráfico de dispersão.
#hue = 'Stage' para colorir nossos pontos de acordo com o estágio de evolução do Pokémon.

sns.lmplot(x='Attack', y='Defense', data=df,
           fit_reg=False,
           hue='Stage', height=10)
```

```
#Agora, digamos que queremos fazer um gráfico de caixa para as estatísticas de combate do nosso Pokémon:
#O gráfico de caixa ou "boxplot" é útil avaliar a distribuição empírica do dados. 
#O boxplot é formado pelo primeiro e terceiro quartil e pela mediana, conforme foi apresentado no minicurso anterior
plt.figure(figsize=(15,8))
sns.boxplot(data=df)
```

Podemos remover algumas colunas como a coluna Total, pois temos estatísticas individuais. Podemos também remover as colunas Stage e Legendary porque não são estatísticas de combate.

Porém, isso não é tão simples de fazer apenas no Seaborn. Em vez disso, é muito mais simples pré-formatar o DataFrame.

```
stats_df = df.drop(['Total', 'Stage', 'Legendary'], axis=1)
plt.figure(figsize=(15,8))
sns.boxplot(data=stats_df)
```

```
#Outra vantagem do Seaborn é que ele vem com temas de estilo
#Os gráficos de violino são visualizações úteis, pois mostram a distribuição dos valores (através da espessura do violino) 
#em vez de apenas as estatísticas resumidas.
#O gráfico de violino é um aprimoramento do boxplot
#A vantagem de gráfico de violino é que ele pode mostrar nuances na distribuição não perceptíveis em um boxplot.

sns.set_style('whitegrid')
plt.figure(figsize=(15,8))
sns.violinplot(x='Type 1', y='Attack', data=df)
```

Como podemos observar, os tipos de fogo tendem a ter estatísticas de ataque mais altas do que os tipos de fantasma, mas também têm uma variância maior.
Agora, os fãs de Pokémon podem achar algo bastante chocante: as cores são absurdas. Por que o tipo Grama é cor de rosa ou o tipo Água é cor de laranja? Devemos consertar isso!

```
pkmn_type_colors = ['#78C850',  # Grass
                    '#8B0000',  # Fire
                    '#6890F0',  # Water
                    '#A8B820',  # Bug
                    '#A8A878',  # Normal
                    '#A040A0',  # Poison
                    '#F8D030',  # Electric
                    '#E0C068',  # Ground
                    '#EE99AC',  # Fairy
                    '#708090',  # Fighting
                    '#F85888',  # Psychic
                    '#B8A038',  # Rock
                    '#705898',  # Ghost
                    '#98D8D8',  # Ice
                    '#7038F8',  # Dragon
                   ]
```

```
plt.figure(figsize=(15,8))
sns.violinplot(x='Type 1', y='Attack', data=df, 
               palette=pkmn_type_colors)
```

Como temos apenas 151 Pokémon em nosso conjunto de dados, será que podemos simplesmente exibir cada ponto? Sim!

Podemos utilizar o gráfico de enxame que mostrará cada ponto, enquanto "empilha" aqueles com valores semelhantes:

```
#O gráfico de enxame ou "swarmplot" é um gráfico de dispersão categórico com pontos não sobrepostos

plt.figure(figsize=(15,8))
sns.swarmplot(x='Type 1', y='Attack', data=df, 
              palette=pkmn_type_colors)
```

Podemos combinar nosso enredo de enxame e o enredo de violino? Afinal, eles exibem informações semelhantes, certo?

```

plt.figure(figsize=(15,8))
 
sns.violinplot(x='Type 1',
               y='Attack', 
               data=df, 
               inner=None, # Remove as barras dentro dos violinos
               palette=pkmn_type_colors)
 
sns.swarmplot(x='Type 1', 
              y='Attack', 
              data=df, 
              color='k', #Tornar os pontos pretos 
              alpha=0.6) #e ligeiramente transparente 
 
plt.title('Ataque por tipo')
```

```
#Vamos concluir este tutorial com algumas visualizações de dados rápidas, apenas para dar uma ideia do que é possível com o Seaborn.

#Calculando correlação
corr = stats_df.corr()
 
#Heatmap
plt.figure(figsize=(15,8))
sns.heatmap(corr)
```

```
#Histogramas permitem representar graficamente as distribuições de variáveis ​​numéricas.
plt.figure(figsize=(15,8))
sns.distplot(df.Attack)
```

```
#Os gráficos de barras ajudam a visualizar as distribuições de variáveis ​​categóricas.

plt.figure(figsize=(15,8))
sns.countplot(x='Type 1', data=df, palette=pkmn_type_colors)

#Rotação eixo-x
plt.xticks(rotation=-45)
```

```
#O "factorplot" facilita a separação de gráficos por classes categóricas

g = sns.factorplot(x='Type 1', 
                   y='Attack', 
                   data=df, 
                   hue='Stage',  #cor por estágio
                   col='Stage',  #separado por estágio
                   kind='swarm', #gráfico de enxame
                   height=5)

#Rotação eixo-x
g.set_xticklabels(rotation=-45)
```

```
#O gráfico de densidade exibe a distribuição entre duas variáveis.
#Os contornos são essencialmente um mapa topográfico para duas variáveis; 
#cada banda de contorno representa uma densidade específica de pontos, aumentando à medida que se aproxima de um pico.

plt.figure(figsize=(15,8))
sns.kdeplot(df.Attack, df.Defense)
```

```
#O "joinplot" combina informações de gráficos de dispersão e 
#histogramas para fornecer informações detalhadas para distribuições bivariadas.

sns.jointplot(x='Attack', y='Defense', data=df,height=10)
```
